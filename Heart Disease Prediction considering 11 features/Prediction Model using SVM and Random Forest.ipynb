```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/heart.csv")
df.head()
```

```python
# removing the outliers of the resting bp column using z-score
df["z_score"] =  (df['RestingBP'] - df['RestingBP'].mean()) / df['RestingBP'].std()
df
```

```python
df.z_score.describe()
```

```python
df = df[df["z_score"].abs() < 3]
df
```

```python
# removing the outliers of the Cholesterol column using z-score

df["z_score"] =  (df['Cholesterol'] - df['Cholesterol'].mean()) / df['Cholesterol'].std()
df = df[df["z_score"].abs() < 3]
df.shape
```

```python
# removing the outliers of the MaxHR column using z-score
df["z_score"] =  (df['MaxHR'] - df['MaxHR'].mean()) / df['MaxHR'].std()
df = df[df["z_score"].abs() < 3]
df.shape
```

```python
# removing the outliers of the Oldpeak column using z-score
df["z_score"] =  (df['Oldpeak'] - df['Oldpeak'].mean()) / df['Oldpeak'].std()
df = df[df["z_score"].abs() < 3]
df.shape
```

```python
df = df.drop(columns=["z_score"])
df
```

```python
# one hot encoding to convert categorical variables to numerical variables
df = pd.get_dummies(df, columns=['ChestPainType', 'Sex', "ExerciseAngina", "RestingECG", "ST_Slope"], drop_first=True)
df
```

```python
# aply scaling to the data by standard scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[["Age", "RestingBP", "Cholesterol", "FastingBS", "MaxHR", "Oldpeak"]] = scaler.fit_transform(df[["Age", "RestingBP", "Cholesterol", "FastingBS", "MaxHR", "Oldpeak"]])
df
```

```python
# using svm to classify the data
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

svm = SVC(kernel='linear', C=1, random_state=42)
x  = df.drop(columns=["HeartDisease"])
y = df["HeartDisease"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
svm.fit(x_train, y_train)
```

```python
svm.score(x_test, y_test)
```

```python
# now we will use bagging to improve the accuracy of the model
from sklearn.ensemble import BaggingClassifier

bag = BaggingClassifier(svm, n_estimators=10, random_state=42, oob_score=True)
bag.fit(x_train, y_train)
```

```python
bag.oob_score_
```

```python
bag.score(x_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(bag, x_train, y_train, cv=5).mean()
```

```python
cross_val_score(svm, x_train, y_train, cv=5).mean()
```

```python
# now we will use decision tree to classify the data
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(random_state=42)
tree.fit(x_train, y_train)
tree.score(x_test, y_test)
```

```python
cross_val_score(tree, x_train, y_train, cv=5).mean()
```

```python
# now we will use bagging to improve the accuracy of the model
bagging_model = BaggingClassifier(tree, n_estimators=10, random_state=42, oob_score=True)
bagging_model.fit(x_train, y_train)
bagging_model.oob_score_
```

```python
cross_val_score(bag, x_test, y_test, cv=5).mean()
```

```python
cross_val_score(bag, x_test, y_test, cv=5).mean()
```

```python
# Now we will determine the best parameters for the two models using grid search
from sklearn.model_selection import GridSearchCV

model_params = {
    "svm" : {
        "model": SVC(),
        "params" : {
            "C" : [0.1, 1, 10, 100],
            "kernel" : ["linear", "rbf", "poly"],
            "gamma" : [0.1, 1, 10, 100, "auto"]
        }
    },
    "tree" : {
        "model": DecisionTreeClassifier(),
        "params" : {
            "max_depth" : [None, 10, 20, 30],
            "min_samples_split" : [2, 5, 10] 
        }
    },
    "bagging" : {
        "model": BaggingClassifier(),
        "params" : {
            "n_estimators" : [10, 50, 100],
            "oob_score" : [True, False],
            "random_state" : [42]
        }
    },
}
```

```python
scores = []
for model_name, mp in model_params.items():
    clf = GridSearchCV(mp["model"], mp["params"], cv=5)
    clf.fit(x_train, y_train)
    scores.append(
        {
            "model" : model_name,
            "best_params" : clf.best_params_,
            "best_score" : clf.best_score_
        }
    )
scores_df = pd.DataFrame(scores)
scores_df
```

```python
# Now we will use random forest to classify the data
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42)
rf.fit(x_train, y_train)
rf.score(x_test, y_test)
```

```python
cross_val_score(rf, x_train, y_train, cv=5).mean()
```

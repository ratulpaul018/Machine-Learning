```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['rainfall'] = le.fit_transform(df['rainfall'])
df
```

```python
x = df.drop(columns=['rainfall', "day"])
y = df['rainfall']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
# determine the model and the parameters to be tuned
from sklearn.ensemble import RandomForestClassifier
# import knn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
scores = []
model_params = {
    'RandomForestClassifier': {
        'model': RandomForestClassifier(),
        'params': {
            'model__n_estimators': [10, 50, 100]
        }
    },
    'KNeighborsClassifier': {
        'model': KNeighborsClassifier(),
        'params': {
            'model__n_neighbors': [3, 5, 7],
            'model__weights': ['uniform', 'distance']
        }
    },
    'LogisticRegression': {
        'model': LogisticRegression(max_iter=10000),
        'params': {
            'model__solver': ['liblinear', 'saga']
        }
    },
    'SVC': {
        'model': SVC(),
        'params': {
            'model__C': [0.1, 1, 10],
            'model__kernel': ['linear', 'rbf']
        }
    }
    
}

for model_name, mp in model_params.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', mp['model'])
    ])
    clf = GridSearchCV(pipe, mp['params'], cv=5, return_train_score=False)
    clf.fit(x_train, y_train)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
df_scores = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
df_scores.sort_values(by='best_score', ascending=False, inplace=True)
```

```python
df_scores
```

```python
clf.best_estimator_
```

```python
model = clf.best_estimator_.fit(x_train, y_train)
model
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(model, x_test, y_test, cv=50).mean()
```

```python
# use bagin on top of the best model
from sklearn.ensemble import BaggingClassifier

bagging_model = BaggingClassifier(model, n_estimators=10, random_state=0)
bagging_model.fit(x_train, y_train)
```

```python
bagging_model.score(x_test, y_test)
```

```python
# apply pca to the data
from sklearn.decomposition import PCA
pca = PCA(0.95)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)
x_train_pca.shape, x_test_pca.shape
```

```python
bagging_model_pca = BaggingClassifier(model, n_estimators=10, random_state=0)
bagging_model_pca.fit(x_train_pca, y_train)
```

```python
bagging_model_pca.score(x_test_pca, y_test)
```

```python
cross_val_score(model, x_test_pca, y_test, cv = 50).mean()
```
